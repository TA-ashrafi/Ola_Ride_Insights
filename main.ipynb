# Data Cleaning Tasks Bengaluru_Ride_Data

```python
import pandas as pd 


df = pd.read_csv('bengaluru_ride_data.csv')

# Removing Duplicate From datset 
df = df.drop_duplicates()

# Fill driver_rating with mean rating per driver_id
df['Driver Ratings'] = df['Driver Ratings'].fillna(df['Driver Ratings'].mean())

# Fill fare_amount (Booking Value) with median fare per city
df['Booking Value'] - df['Booking Value'].fillna(df['Booking Value'].median())


# 4. Convert trip_start_time & trip_end_time into proper datetime format
df['trip_start_time'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d-%m-%Y %H:%M')

df['trip_end_time'] = df['trip_start_time'] 

df = df[df['Booking Value'] >= 0]

# 6. Extract date, day_of_week, and hour from trip_start_time

df['Date'] = df['trip_start_time'].dt.date
df['Days_of_week'] = df['trip_start_time'].dt.day_name()
df['Hour']  = df['trip_start_time'].dt.hour

# 7. Normalize pickup_location & drop_location
# Removeing extra spaces and standardize cas
df['Pickup Location'] = df['Pickup Location'].str.strip().str.title()
df['Drop Location'] = df['Drop Location'].str.strip().str.title()

# Arranging Columns
arranged_columns = [
    'Booking Id', 
    'Customer Id', 
    'Vehicle Type', 
    'Pickup Location', 
    'Drop Location',
    'Trip Start Time', 
    'Trip End Time', 
    'Date', 
    'Days Of Week', 
    'Hour',
    'Booking Value', 
    'Ride Distance', 
    'Driver Ratings', 
    'Customer Ratings',
    'Avg Vtat', 
    'Avg Ctat', 
    'Booking Status', 
    'Cancelled Rides By Customer',
    'Reason For Cancelling By Customer', 
    'Cancelled Rides By Driver',
    'Reason For Cancelling By Driver', 
    'Incomplete Rides', 
    'Incomplete Rides Reason'
]

column_mapping = {
    'Trip_Start_Time': 'Trip Start Time',
    'Trip_End_Time': 'Trip End Time',
    'Days_Of_Week': 'Days Of Week'
}

# Renaming Columns
df.columns = [col.title() for col in df.columns]
df.columns = [column_mapping.get(col, col) for col in df.columns]
df = df[arranged_columns]

# EXporting Cleaned Data
# df.to_csv('cleaned_bengaluru_ride_data.csv', index=False)
```
